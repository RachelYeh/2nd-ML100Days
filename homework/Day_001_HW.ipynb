1. 核心問題為何 (tips：如何定義 「提升業績 & 你的假設」)

我認為「媒合(Matching)」這個動作是相關關鍵的，若無刻意去促成媒合，基本上很難達成交易。畢竟車子在外隨機亂跑，要能剛好遇到一名有需求的乘客，是得碰運氣的。所以首先需要一個「媒合平台」，讓供需有交集之處。

接下來再探討應如何提升業績：每一次的載客，我們都會希望車程能拉越長越好，因為行駛里程數越多、賺的錢越多。此外，每次的載客之間的空檔必須要能最小化，讓司機都能很快就接到客人，才能減少成本。所以「單趟車程越久」以及「能在短時間內載到乘客」是相當重要的評估點。以下針對這兩點去探討：

－「單趟車程越久」：媒合平台中乘客須寫下起始出發地點及最終目的地，讓系統可估算出車程距離後，車程愈遠司機所獲得車資也越高，因此選擇服務該乘客的獲益也會越高。
－「能在短時間內找到下一位客人」：承上，就算有名乘客的車程價值較高，但若他所在位置與司機距離較遠，為了去載他反而可能耗費過高的移動成本。因此平台另需計算移動成本，「載客獲益」扣除掉「移動成本」的最高者，才最應優先推薦給司機。

值得注意的地方是，若目的地為相當偏遠荒涼的地方，即使單趟車程久，卻可能造成將使得司機難以找到下一位乘客，所以也應考慮抵達「目的地易接客的程度」，可透過設定權重的方式來影響系統的推薦排名，像是機場、車站等，旅客來來往往方便載客，權重應設為最高，越偏離市中心處權重應設愈低。而當車隊經營久了，自然也會有載客熱區的紀錄，可依此調整系統。

另外，研究「乘客特徵(profiling)」也有助於提升對客群的瞭解，例如怎樣的年齡層較常搭車、搭乘較遠距離的多為怎樣的乘客等，甚至是他們的偏好需求(後車箱空間大、採預約還是現叫方式、是否用行動支付、最久願意等待時間等)，這些都能幫助我們更精準掌握客群，能懂得從何面向去提升服務品質，也可釋出針對性的促銷優惠訊息，吸引特定客群的注意。
